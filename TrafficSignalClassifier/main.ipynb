```python
import tensorflow as tf
from tensorflow.keras import layers,models
import numpy as np
import cv2
import os
import yaml
from sklearn.model_selection import train_test_split
```

```python
def load_data(path):
    with open(path,'r') as f:
        return yaml.safe_load(f)
    
```

```python
test_meta=load_data('C:\\Users\\grace\\OneDrive\\Desktop\\vscode\\dl\\DeepLearningProjects\\TrafficSignalClassifier\\data\\1\\test.yaml')
train_meta=load_data('C:\\Users\\grace\\OneDrive\\Desktop\\vscode\\dl\\DeepLearningProjects\\TrafficSignalClassifier\\data\\1\\train.yaml')
```

```python
def load_images(metadat,base_dir):
    images=[]
    labels=[]
    class_map={'Red':0,'Green':1,'Yellow':2,'red':0,'green':1,'yellow':2}
    for entry in metadat:
        img_path=os.path.join(base_dir,os.path.basename(entry['path']))
        img=cv2.imread(img_path)
        if img is None:
            print(f'Could not read image {img_path}')
            continue
        if len(entry['boxes'])>0:
            label=entry['boxes'][0]['label']
            img=cv2.resize(img,(64,64))/255.0
            images.append(img)
            labels.append(class_map[label])
        else:
            print(f"No boxes found in {img_path}")
    return np.array(images),np.array(labels)

x_train,y_train=load_images(train_meta,'C:\\Users\\grace\\OneDrive\\Desktop\\vscode\\dl\\DeepLearningProjects\\TrafficSignalClassifier\\data\\1\\rgb\\train')
x_test,y_test=load_images(test_meta,'C:\\Users\\grace\\OneDrive\\Desktop\\vscode\\dl\\DeepLearningProjects\\TrafficSignalClassifier\\data\\1\\rgb\\test')
```
